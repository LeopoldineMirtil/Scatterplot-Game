# Scatter Plot Game

```python
import matplotlib.pyplot as plt
import numpy as np
import random

#initialize count of score and rounds
score = 0 
round = 3

#generate points at random
x = random.sample(range(-14, 14), 3)
y = random.sample(range(-14, 14), 3)

# create coordinates
co1 = [x[0], y[0]]
co2 = [x[1], y[1]]
co3 = [x[2], y[2]]


##create graph for coordinates
xmin = -15
xmax = 15
ymin = -15
ymax = 15

fig, ax = plt.subplots()
plt.axis([xmin,xmax,ymin,ymax]) # window size
plt.plot([xmin,xmax],[0,0],'gray') # x axis
plt.plot([0,0],[ymin,ymax],'gray') # y axis

plt.plot(x, y, 'go') # plot points
plt.grid()
plt.show()


while True:
    #input coordinates
    guess = input("Enter your guess coordinates, with 'x' and 'y' separated by a comma: ")
    g_s = guess.split(",") 
    g = [int(g_s[0]), int(g_s[1])] 

    #compare guesses
    if g == co1:
        score += 1
        round -= 1
        print(f'Correct!') 
    elif g == co2: 
        score += 1
        round -= 1
        print(f'Correct!') 
    elif g == co3:
        score += 1
        round -= 1
        print(f'Correct!') 
    else: 
        round -= 1
        print(f'Sorry, wrong guess')

    if round ==0:
        break

print("\n", "*****Game Over*****")
if score == 3:
    print(f"Great job! You scored {score} out of 3")
elif score == 2:
    print(f"Not bad! You scored {score} out of 3")
elif score == 1: 
    print(f"Better luck next time. You scored {score} out of 3")  
else:
    print(f"Whoof...You scored {score} out of 3")
```
